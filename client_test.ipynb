```python
def read_in_chunks(file_object, chunk_size=1024):
    """Lazy function (generator) to read a file piece by piece.
    Default chunk size: 1k."""
    while True:
        data = file_object.read(chunk_size)
        print (data)
        b_data.append(data)
        if not data:
            break
        yield data
```

```python
# 클라이언트 측에서 음성 파일을 열고 청크 단위로 서버에 전송
file_path = 'upload_audio/hello_chatgpt.wav'  # 여기에 테스트할 음성 파일 경로를 입력하세요
url = "https://sheepdog-bold-bulldog.ngrok-free.app/speak2speak_chunk"  # FastAPI 서버 엔드포인트 URL
```

```python
### 테스트용 chunk 쪼개서 받고 chunk 쪼개서 보내기 함수 테스트용

# 스트리밍 요청 보내기
with open(file_path, 'rb') as f:
    response = requests.post(
        url, 
        data=read_in_chunks(f), 
        headers={'Content-Type': 'application/octet-stream'},
        stream=True  # 응답을 스트리밍 방식으로 받도록 설정
    )
    print("완료")

# 서버 응답 출력 (오디오 파일로 저장하기)
with open('response_audio.mp3', 'wb') as output_file:
    # 응답을 청크 단위로 읽기
    for chunk in response.iter_content(chunk_size=1024):
        if chunk:  # 청크가 있을 경우에만
            output_file.write(chunk)  # 청크를 파일에 쓰기
            print(f"청크 크기: {len(chunk)} bytes")  # 청크 크기를 출력

print("음성 변환 완료: response_audio.mp3 파일로 저장됨.")
```

```python
### 메타 데이터를 같이 동봉해서 보내는 방법에 클라이언트 테스트 코드인데 이거 너무 복잡해서 유기함

import requests

# 스트리밍 데이터 가져오기
response = requests.post("https://sheepdog-bold-bulldog.ngrok-free.app/chatbot_test/from_vec", json={"romm_id": 1}, stream=True)

# 응답 데이터 처리
metadata_received = False
audio_data = b''

for chunk in response.iter_content(chunk_size=1024):
    if not metadata_received:
        try:
            # 첫 번째 청크에서 메타데이터를 읽고 나머지는 오디오 데이터로 처리
            data = chunk.decode('utf-8')
            if "\n\n" in data:
                metadata, audio_chunk = data.split("\n\n", 1)
                print("Metadata:", metadata)  # 메타데이터 출력
                metadata_received = True
                audio_data += audio_chunk.encode('utf-8')  # 남은 부분을 오디오 데이터로 추가
            else:
                continue
        except UnicodeDecodeError:
            # 만약 디코딩 에러가 발생하면 이는 오디오 데이터의 일부이므로 오디오 데이터를 추가
            audio_data += chunk
            metadata_received = True  # 첫 번째 메타데이터 부분을 넘어갔음을 표시
    else:
        # 이후 청크들은 오디오 데이터로 처리
        audio_data += chunk

# 오디오 데이터를 파일로 저장
with open("output.mp3", "wb") as audio_file:
    audio_file.write(audio_data)

print("오디오 파일이 저장되었습니다.")
```

```python
import requests

def test_speak2speak_multipart(file_path):
    # 오디오 파일을 바이너리 모드로 열기
    with open(file_path, "rb") as f:
        # 파일을 포함한 POST 요청 보내기
        response = requests.post(
            "http://127.0.0.1:8080/speak2speak_multipart",  # 실제 엔드포인트 URL로 교체하세요
            files={"audio_file": ("test_audio.wav", f, "audio/wav")}  # MIME 타입을 파일 형식에 맞게 조정하세요
        )
        # 응답 상태 코드 확인
        if response.status_code == 200:
            # 스트리밍된 오디오 데이터를 파일로 저장
            with open("output_audio.mp3", "wb") as output_file:
                for chunk in response.iter_content(chunk_size=1024):
                    output_file.write(chunk)
            print("오디오 파일을 받아서 output_audio.mp3로 저장했습니다.")
        else:
            print(f"오디오를 받는 데 실패했습니다. 상태 코드: {response.status_code}")

# 테스트할 오디오 파일 경로를 입력하세요
file_path = "upload_audio/a_speak.mp3"

# 테스트 함수 실행
test_speak2speak_multipart(file_path)
```
