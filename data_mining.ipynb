```python
df = pd.read_csv("종이책_독서율_성인.csv",encoding="cp949")
df.describe
```

```python
df_read = pd.read_csv("종이책_독서율__읽음.csv",encoding = "cp949")
df_new = df_read.iloc[1]
df_new_values = df_new[2:]
df_new_values
df_new_values = df_new_values.astype(float)
```

```python
plt.title("연도별 '독서 안함' 지수")
plt.bar(df_values.index,df_values.values)
plt.ylim(0)  # y축의 상단도 데이터 최대값보다 약간 크게 설정

plt.show()
```
